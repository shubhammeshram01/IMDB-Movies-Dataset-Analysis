## 19) Classify movies based on rating [Excellent, Good, Average]

```python
data.head(2)
```

```python
def rating(rating):
    if rating >= 7.5:
        return "Excellent"
    elif rating >= 6.5:
        return "Good"
    else:
        return "Average"
```

```python
data["Rating_Class"] = data["Rating"].apply(rating)
```

```python
data.head(5)[["Rating" , "Rating_Class"]]
```

## 21) Find unique values from Genre

```python
data["Genre"]
```

```python
data["Genre"].unique()
```

```python
list_1 = []

for i in data["Genre"]:
    list_1.append(i.split(","))
    
```

```python
list_1
```

```python
list_2 = []

for item in list_1:
    for item1 in item:
        list_2.append(item1)   
```

```python
list_2
```

```python
unique_list = []

for i in list_2:
    if i not in unique_list :
        unique_list.append(i)
```

```python
unique_list
```

```python
len(unique_list)
```
